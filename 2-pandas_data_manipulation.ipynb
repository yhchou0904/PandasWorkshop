## Data Type
pandas所帶的原生資料型態包含以下兩種：
1. series
2. DataFrame

### Series
* array-like data type
* we can set index for every element

````python
my_series = pd.Series([4, 7, -5, 3])
print(my_series)
````

````python
my_series = pd.Series([4, 7, -5, 3], index=['a', 'b', 'c', 'd'])
print(my_series)
````

````python
dt = {'a':1, 'b':2, 'c':3, 'd':4}
my_series = pd.Series(dt)
print(my_series)
````

### Data Frame
* usually 2-D data structure
* we can specify *row*, *column*, *index*

````python
dt = {'a':[1, 5], 'b':[2, 6], 'c':[3, 7], 'd':[4, 8]}
my_df = pd.DataFrame(dt)
my_df.index = ['一', '二']
my_df
````

### Filling N/A
We can use `DataFrame.fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None, **kwargs)` to fill N/A.
* value: specifing what to fill in every column/ row
* method: how to fill, including ‘backfill’, ‘bfill’(fill with next value), ‘pad’, ‘ffill’(fill with last value)
* limit: how many N/A should fill with values/ ways we decide

````python
f = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, 5],
                   [np.nan, 3, np.nan, 4]],
                   columns=list('ABCD'))
f
````

````python
f.fillna(7)
````

````python
f.fillna(value={'A':7, 'B':6, 'C':8})
````

````python
f.fillna(method='ffill', limit=1)
````

### How to fill with mean?

````python
f.fillna(f.mean())
````

## Merge, Join, Concatenate
We can join dataframes by several methods:
* concatenate: <br>
```
pandas.concat(objs, axis=0, join='outer', ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=True)
```
* merge: <br>
```
pandas.merge(left, right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)
```
* join: <br>
```
DataFrame.join(other, on=None, how='left', lsuffix='', rsuffix='', sort=False)
```

In the sample, we have at most 8 people's income data in the same house:

````python
df_105_itm = df_105[[i for i in df_105.columns if ('itm40' in i) and (i != 'itm400') and (i != 'itm409')]]
df_105_itm
````

````python
tmp_105_without_methods = tmp_105_b[:]
````

## Apply
We can do same transformation to every single value in a column

````python
inc_data_105['INC'].head()
````

````python
inc_data_105['INC_log'] = inc_data_105['INC'].apply(lambda x: math.log(x))
inc_data_105['INC'].head()
````

Or by using self-defined function

````python
def generatingHead(row):
    if row['b2'] == '本人':
        row['HEAD'] = 1
    else:
        row['HEAD'] = 0
    return row
````

````python
inc_data_105 = inc_data_105.apply(lambda x: generatingHead(x), axis=1, result_type='expand')
inc_data_105
````

````python
def marriage_status(row):
    try:
        if row['b16'][:2] == '配偶':
            row['b16_已婚'] = 1
        elif row['b16'] == '未婚':
            row['b16_未婚'] = 1
        else:
            row['b16_其他'] = 1
        return row
    except:
        return row
````

````python
inc_data_105 = inc_data_105.apply(lambda x: marriage_status(x), axis=1, result_type='expand')
````

````python
inc_data_105 = inc_data_105.drop(['b16'], axis=1)
````

## Rename Columns

````python
inc_data_105.rename(columns={'b13':'WORKING', 'b14':'EARNING', 'b3':'GENDER', 'b4':'AGE', 'b2':'HH_STATUS'}, inplace=True)
inc_data_105
````
